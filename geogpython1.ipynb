```python
## Task 13
def countdown(num):
    count = num
    while(count != 0):
        print("T-minus " + str(count))
        count-= 1
countdown(15)


## Task 14 
def quartile(num): 
    if(num >= 0 and num<= 25):
        print("1st Quartile")
    elif(num > 25 and num <= 50):
        print("2nd Quartile")
    elif(num > 50 and num <= 75):
        print("3rd Quartile")
    elif(num > 75 and num <= 100):
        print("4th Quartile")
    elif(num < 0 or num > 100):
        print("Error")

quartile(13)
quartile(32)
quartile(69)
quartile(95)
quartile(-16)
quartile(213)
```
